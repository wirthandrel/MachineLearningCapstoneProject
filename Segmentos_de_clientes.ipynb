```python
# TODO: Escalone os dados utilizando o algoritmo natural
log_data = np.log(data)

# TODO: Escalone a amostra de dados utilizando o algoritmo natural
log_samples = np.log(samples)

# Produza uma matriz de dispers칚o para cada par de atributos novos-transformados
pd.plotting.scatter_matrix(log_data, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
print("Destribui칞칚o dos dados ap칩s transforma칞칚o logaritmica")
```

```python
outliers_list=[]
outliers_repeated=[] 
# Para cada atributo encontre os pontos de dados com m치ximos valores altos e baixos
for feature in log_data.keys():
    #print("feature: {}".format(feature))
    # TODO: Calcule Q1 (25췈 percentil dos dados) para o atributo dado
    Q1 = np.percentile(log_data[feature], 25)
    
    # TODO: Calcule Q3 (75췈 percentil dos dados) para o atributo dado
    Q3 = np.percentile(log_data[feature], 75)
    
    # TODO: Utilize a amplitude interquartil para calcular o passo do discrepante (1,5 vezes a varia칞칚o interquartil)
    step = (Q3 - Q1) * 1.5
    
    # Mostre os discrepantes
    #print("Data points considered outliers for the feature '{}':".format(feature))
    outliers_log_data = log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))]
    #display(outliers_log_data)
   
    for outlier in outliers_log_data.index:
        if outlier not in outliers_list:
            outliers_list.append(outlier)
        elif outlier not in outliers_repeated:
            outliers_repeated.append(outlier)    
    
# OPCIONAL: Selecione os 칤ndices dos pontos de dados que voc칡 deseja remover
outliers  = outliers_list + outliers_repeated

# Remova os valores aqui, caso nenhum tenha sido especificado
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)


#print("outliers duplicados {}".format(outliers_repeated))

# TODO: Escalone os dados utilizando o algoritmo natural
log_data = np.log(data)

# TODO: Escalone a amostra de dados utilizando o algoritmo natural
log_samples = np.log(samples)

# Produza uma matriz de dispers칚o para cada par de atributos novos-transformados
pd.plotting.scatter_matrix(log_data, alpha = 0.3, figsize = (14,8), diagonal = 'kde');
print("Destribui칞칚o dos dados ap칩s remo칞칚o dos outliers e transforma칞칚o logaritmica")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
best_score = 0
best_cluster_number = 0
for cluster_number in np.arange(2, 24, 3):

    # TODO: Aplique o algoritmo de clustering de sua escolha aos dados reduzidos 
    clusterer = KMeans(n_clusters=cluster_number).fit(log_data)

    # TODO: Preveja o cluster para cada ponto de dado
    preds = clusterer.predict(log_data)

    # TODO: Ache os centros do cluster
    centers = clusterer.cluster_centers_

    # TODO: Preveja o cluster para cada amostra de pontos de dado transformados
    sample_preds = clusterer.predict(log_samples)

    # TODO: Calcule a m칠dia do coeficiente de silhueta para o n칰mero de clusters escolhidos
    score = silhouette_score(log_data, preds)
    
    #print("Numero de cluster {} score {}".format(cluster_number, score))
    
    if(best_score < score):
        best_score = score
        best_cluster_number = cluster_number
        
print("Melhor score: {} numero de clusters {}".format(best_score, best_cluster_number))
```
